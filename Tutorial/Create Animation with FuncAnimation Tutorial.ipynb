```python
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
import math         

def plot_curves(frame, *kwargs):
    ax.clear()
    # the FuncAnimation cycles through each frame in frames,
    
    df[:frame+1].plot.line(linewidth = linewidth, ax = ax,legend = False)
    # Turn the text on the x-axis so that it reads vertically
    ax.set_xlim(left = 0, right = max_x)
    ax.set_ylim(bottom = 0, top = max_y)
    ax.tick_params(axis='x', rotation=90)
    ax.set_title("Grow a Parabola", fontsize = 70)
    print(frame)
    
def init(*kwargs):
    # Get rid of tick lines perpendicular to the axis for aesthetic
    ax.tick_params('both', length=0, which='both')
    #plt.xticks([i for i in range(len(data.index))], list(data.index))
    ax.tick_params(axis='x', rotation=90)
    # transform y-axis values from sci notation to integers
    ax.set_xlim(left = 0, right = max_x)
    ax.set_ylim(bottom = 0, top = max_y)
    #vals = ax.get_yticks()
    #ax.set_yticklabels([round(y,2) for y in vals])

list_len = 201
parabola = [(x - 100) ** 2 for x in range(list_len)]

max_y = max(parabola)
max_x = list_len
frames = [i for i in range(list_len)]

df = pd.DataFrame(parabola, index = frames)
fig, ax = plt.subplots(figsize=(40,20))   
plt.rcParams.update({"font.size": 30})
linewidth = 5
# the functions will unpack the tuple. The same names variable names
# are used in the function
kwargs = (df, fig, ax, list_len, max_x, max_y, linewidth)
anim = FuncAnimation(fig, plot_curves, frames = frames, 
                     blit = False, init_func = init, interval=25, 
                     fargs =kwargs)
# Use the next line to save the video as an MP4.
anim.save("GrowAParabola.mp4", writer = "ffmpeg")
plt.close()
```
