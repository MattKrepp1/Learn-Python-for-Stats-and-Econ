## Monte Carlo Simulation

### 1. Build list of points drawn from a standard normal distribution

```python
random_list = []
sim_length = 1000
mean = 0
sigma = .5

for i in range(sim_length):
    random_list.append(random.normalvariate(mean,sigma))
plt.plot(random_list, ls = "", marker = ".", markersize = 2)
plt.show()
plt.close()
```

## 2. Take the cumulative sum of the list at each index to create a randomly generated line

```python
simulation = []

for i in range(len(random_list)):
    val = random_list[i]
    if i == 0:
        simulation.append(val)
    else:
        simulation.append(simulation[i - 1] + val)
plt.plot(simulation)
plt.show()
plt.close()
    
```

### Comparing randomly generate points and the line it generates

```python
print("per\tline\t\t\tpoints")
list(zip(np.arange(len(simulation)),simulation, random_list))
```

# 3. Create multiple simulations. Record each result in a dictionary, with each period of the simulation being represented in the second layer of the dictionary-of-dictionaries

```python
monte_carlo_sim_dict = {}
num_sims = 2000
periods = 1000
for i in range(num_sims):
    monte_carlo_sim_dict[i] = {}
    for period in range(periods):
        # create shorthand variable for current simulation
        curr_sim = monte_carlo_sim_dict[i]
        if period == 0:
            curr_sim[period] = random.normalvariate(mean, sigma)
        else:
            curr_sim[period] = curr_sim[period - 1] + random.normalvariate(mean, sigma)
```

### Transform to dataframe

```python
monte_carlo_sim_df = pd.DataFrame(monte_carlo_sim_dict)
monte_carlo_sim_df
```
