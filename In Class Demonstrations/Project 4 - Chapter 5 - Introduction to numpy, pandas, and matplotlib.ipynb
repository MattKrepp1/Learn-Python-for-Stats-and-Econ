## forming an array with a list of different types of values

In the last case, float_array, we check to see if the transformation of a list of all floats is the same as the transformation of a mixed list of float and ints.

```python
int_float_list = [1,2,3,4., 5]
print("int_float_list")
for val in int_float_list:
    print(val, type(val))
mystery_array = np.array(int_float_list)
print()

print("mystery_array")
for val in mystery_array:
    print(val, type(val))
print()

print("float_array")
float_list = [1.,2.,3.,4.,5.]
float_array = np.array(float_list)
for val in float_array:
    print(val, type(val))
```

# Create random data for a dataframe

```python
macro_dict = {"GDP":{},
            "Money":{},
#            "Real GDP":{},
            "Price Level":{}}
```

```python
for key in macro_dict:
    for i in range(1990, 2018):
        macro_dict[key][i] = np.random.random() #* 10000
```

```python
macro_dict
```

```python
macro_df = pd.DataFrame(macro_dict)
macro_df
```

## Create new column in dataframe using values from other df columns

```python
macro_df["Velocity"] = macro_df["GDP"].div(
    macro_df["Money"])
macro_df["Real GDP"] = macro_df["GDP"].div(
    macro_df["Price Level"])
macro_df
```

```python
macro_df.loc[1995:2000]
```

## You can call a subset of rows by passing a list of keys

```python
macro_df[["Real GDP", "Money"]]
```

```python
macro_df.loc[1995:2000, ["Real GDP", "Money"]]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
macro_df.plot.line()
```

```python
for key in macro_df:
    # fig, ax = plt.subplots() allows you to adjust different features of the plot box
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line(ax = ax)
    
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```

```python
for key in macro_df:
    # fig, ax = plt.subplots() allows you to adjust different features of the plot box
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line(color = "C5", ax = ax)
    
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```

```python
for key in macro_df:
    # fig, ax = plt.subplots() allows you to adjust different features of the plot box
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line(cmap = "viridis", ax = ax)
    macro_df[key].plot.line(ls= "", marker = ".", ax = ax)
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```

```python
for key in macro_df:
    # fig, ax = plt.subplots() allows you to adjust different features of the plot box
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line(color= "C0", ax = ax)
    macro_df[key].plot.line(ls= "", marker = ".", color = "C1", ax = ax)
    # print y-value for each date
    for ix, val in zip(macro_df.index, macro_df[key]):
        ax.text(ix, val, str(round(val,2)), color = "C2")
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```
