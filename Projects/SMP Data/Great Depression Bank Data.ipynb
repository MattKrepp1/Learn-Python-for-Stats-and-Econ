```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel("BankData.xlsx", index_col = [0], parse_dates = True)
data
```

```python
import datetime
def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_codes.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            # If dataframe already exists, add new column
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()
            
    return df
```

```python
import pandas_datareader.data as web
data_dict = {"Dow Jones":"M1109BUSM293NNBR"}
freq = "M"
start = datetime.datetime(1929, 1, 1)
end = datetime.datetime(1932, 12, 31)
# end =datetime.datetime.today()

djia = gather_data(data_dict, start, 
          end = end, freq = freq)
```

```python
data["Dow Jones"] = djia["Dow Jones"].values
data["Total Loans and Investments\n(No Broker Loans)"] = data["Total Loans and Investments"].sub(data["Total Broker Loans"])
```

```python
norm = data.copy()
for key in norm:
    norm[key] = norm[key].div(norm[key].iloc[9])
norm
```
